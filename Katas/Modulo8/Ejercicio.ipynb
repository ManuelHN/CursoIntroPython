```python
planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}
```

```python
moons = planet_moons.values()
planets = len(planet_moons.keys())
print(moons)
print(planets)
```

```python
total_moons = 0
for value in moons:
    total_moons = total_moons + value
average = total_moons / planets
print(f'Existen un total de {total_moons} lunas en el sistema')
print(f'En promedio hay {round(average, 2)} lunas por planeta')
```
